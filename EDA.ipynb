# Conclusions

The training dataset contains 77.9% of images without ships, which looks like an overestimated number. 

It is worth testing the model training option with a proportion of 50% images without boxes and 50% images with boxes, that is, in the training data set, reduce the number of images without ships from 150,000 to 42,556 (42,556 - number of unique images with ships; or 81723 entries with ship masks). 

**As a result, the training dataset will be reduced more than two times (for unique images):**  
    - from 192,556 (150,000 + 42,556) to 85,112 unique images;  
    - or from 231,723 to 124.279 entries with masks.

# Main observations

* Ship count distribution on Train Data:
    * 150.000 images (77%) do not contain ships ;
    * 42.556 images (23%) contain from 1 ship (14.08%) to 15 ships (0.03%).

* All images in Train and Test sets have the same size: (768, 768)

## Visualization image + masks

```python
# ref: https://www.kaggle.com/paulorzp/run-length-encode-and-decode
def rle_decode(mask_rle, shape=(768, 768)):
    '''
    mask_rle: run-length as string formated (start length)
    shape: (height,width) of array to return 
    Returns numpy array, 1 - mask, 0 - background

    '''
    s = mask_rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape).T  # Needed to align to RLE direction
```

```python
# Choose one image
ImageId = '0005d01c8.jpg'
img = imread(os.path.join(TRAIN_PATH, ImageId))
img_masks = df_train['EncodedPixels'][df_train['ImageId'] == ImageId].tolist()

# Take the individual ship masks and create a single mask array for all ships
all_masks = np.zeros((768, 768))
for mask in img_masks:
    all_masks += rle_decode(mask)

# Visualization
fig, axarr = plt.subplots(1, 3, figsize=(15, 40))
axarr[0].axis('off')
axarr[1].axis('off')
axarr[2].axis('off')
axarr[0].imshow(img)
axarr[1].imshow(all_masks)
axarr[2].imshow(img)
axarr[2].imshow(all_masks, alpha=0.3)
axarr[0].set_title('initial image')
axarr[1].set_title('ship masks')
axarr[2].set_title('initial image with ship masks')
plt.tight_layout(h_pad=0.1, w_pad=0.1)
plt.show()
```

## Short Train DF: 50%_with_ships + 50%_without_ships

```python
# Number of unique images with ship
n_samples = df_train.ImageId[df_train.EncodedPixels.notnull()].nunique()

# Empty flag for filters
df_train['flag'] = 0

# Add flag=1 for images with ship
df_train['flag'][df_train.EncodedPixels.notnull()] = 1

# List indexes of random 42556 images among 150.000 images without ships
index_samples = df_train[df_train.EncodedPixels.isnull()].sample(n_samples).index

# Add flag=1 for 42556 random images without ships
df_train['flag'][df_train.index.isin(index_samples)] = 1

# Short DF with 50% unique images with ship and 50% images without ship
df_short_train = df_train[['ImageId', 'EncodedPixels']][df_train.flag == 1]
print('Unique images with ship: ', df_short_train.ImageId[df_train.EncodedPixels.notnull()].nunique())
print('Unique images without ship: ', df_short_train.ImageId[df_train.EncodedPixels.isnull()].nunique())
```
